**Используемые функции**

```python
def base_check(data: pd.DataFrame, 
               name_column: str,
               is_bar: bool = False,
               hist_bins: int = 0, 
               plt_rot: int = 0) -> None:
    if any(data[name_column].isna()):
        print(f"Количество пропущенных значений {data[name_column].isna().sum()}")
        print(f"Процент пропущенных значений {data[name_column].isna().mean():.2f}%")
    describ = data[name_column].agg(['min','median','max'])
    if hist_bins == 0:
        hist_bins = int(describ['max'] - describ['min'] + 1)
    if is_bar:
        data[name_column].value_counts().sort_index().plot.bar(rot=plt_rot)
    else:
        data[name_column].hist(bins=hist_bins, xrot=plt_rot)
    plt.title(title_dict.get(name_column, name_column))    
    plt.show()
    print(describ)
```

#### Высота потолков (м) `ceiling_height`

```python
base_check(df_realty, 'ceiling_height', hist_bins=100)
```

**38.8%** пропущенных значений. Можно предположить, что параметр не заполнялся в тех случаях, когда нет данных или когда высота потолка значимо не отличается от "средней" принятой высоты потолков в квартирах. **Заменим** пропущенные значения на медианные.

```python
df_realty['ceiling_height'] = df_realty['ceiling_height'].fillna(df_realty['ceiling_height'].median())
```

По графику видны выборсы. Сделаем предположение, что минимально допустимая высота полотков 2м, максимально допустимая - 5м (такие квартиры - редкость, но встречаются, например, студии для двухуровневой планировки или старые дома).

Рассмотрим все, что не попадает в эти значения:

```python
df_realty.query('ceiling_height > 5 or ceiling_height < 2')['ceiling_height'].sort_values().unique()
```

Значения от 20 до 32 - можно предположить, что допущена ошибка при вводе данных и неправильно отделена дробная часть. 

```python
df_realty.loc[(df_realty['ceiling_height'] >= 20) & (df_realty['ceiling_height'] <= 32),'ceiling_height'] /= 10
```

Обновим список аномальных значений:

```python
ceiling_height_anomaly = df_realty.query('ceiling_height > 5 or ceiling_height < 2')['ceiling_height']
ceiling_height_anomaly.sort_values().unique()
```

Посмотрим информацию по квартирам, которые попадают в этот список:

```python
df_realty.query('ceiling_height in @ceiling_height_anomaly').pivot_table(index='ceiling_height',
                                                                         values=['floor','rooms','total_area'],
                                                                         aggfunc='median')
```

Нет очевидной связи между высотой потолков `'ceiling_height'` и другими параметрами квартиры, такими как: общая площадь, количество комнат, этаж. Например, высота полотка более 5м не означает, что площадь квартиры будет большой (для значения **5.8** площадь **71**)


Сделаем предположение, что была допущена ошибка при вводе данных и заменим все аномальные значения на медианное.

```python
df_realty.loc[df_realty['ceiling_height']
              .isin(ceiling_height_anomaly),'ceiling_height'] = df_realty['ceiling_height'].median()
```

И построим диаграму снова:

```python
base_check(df_realty, 'ceiling_height')
```

Аномальных значений нет. Предобработка данных столбца `'ceiling_height'` завершена.

Предположение опровергнуто.<br>

Заполнить пропущенные значения сложно - медианное или среднее значение не походят из-за того, что общее количество этажей должно быть не меньше, чем этаж квартиры. Заполнить общее количество этажей в доме на этаж квартиры - может исказить данные для оценки последнего этажа.


Т.к. этаж может влиять на привлекательность квартиры (находится на первом, последнем или нет), а эти данные невозможно получить при пропущенном значении максимального количества этажей и количество таких объектов небольшое **86** объектов -  удалим данные с пропущенными `floors_total`

####  Название населённого пункта `locality_name`

Оценим количество пропущенных значений.

```python
df_realty['locality_name'].isna().sum()
```

Для названий населенных пунктов нельзя подобрать замену для пропущенных значений и данный параметр важен для исследования, т.к. квартиры с одинаковыми параметрами в зависимости от населенного пункта (например, крупный город и маленькая деревушка) могут значительно отличаться по стоимости и популярности. Удалим строки с пропущенными `'locality_name'`

```python
df_realty = df_realty[df_realty['locality_name'].notna()]
```

Далее, попробуем найти неявные дубликаты:

```python
df_realty['locality_name'].unique()
```

```python
len(df_realty['locality_name'].unique())
```

Первоначально стоит обработать наиболее частую ошибку при строковых данных - замена ё на е. Т.к. в данном случае (365 разнообразных наименований) выборка достаточно большая, для экономии времени и дальнейшей обработки данных имеет смысл везде заменить ё на е.

```python
df_realty['locality_name'] = df_realty['locality_name'].str.replace('ё','е')
```

Посмотрим, какие типы населенных пунктов встречаются в выборке:
- уберем все названия из одного слова
- для остальных удалим последнее слово из названия
<br/>

Это даст грубую выборку с погрешностью, если название населенного состоит из 2 и более слов.

```python
df_realty_type = df_realty['locality_name'].str.replace(r'(^|[ ])\S*?$','',regex=True)
df_realty_type.unique()
```

Можно выделить три крупных типа: "деревня", "поселок", "садовое товарищество". Рассмотрим каждое из них и подготовим списки для замены:

```python
df_realty_type[df_realty_type.str.contains('посел')].unique()
```

Сделаем допущение, `'городской поселок'`, `'поселок городского типа'`, `'поселок станции'`, `'поселок при железнодорожной станции'` и `'коттеджный поселок'` можно заменить на `'поселок'`

```python
df_realty_type[df_realty_type.str.contains('дере')].unique()
```

Для `деревня` нет необходимости в заменах, т.к. разница обусловлена названием деревни в 2 слова

```python
df_realty_type[df_realty_type.str.contains('садо')].unique()
```

Сделаем допущение, что `'садоводческое некоммерческое товарищество'` можно заменить на `'садовое товарищество'`

```python
df_realty['locality_name'] = (
                                df_realty['locality_name']
                                .str.replace(
                                    r'(коттеджный |городской )?поселок( станции)?( городского типа)?( при железнодорожной станции)?',
                                    'поселок', 
                                    regex=True)
)

df_realty['locality_name'] = (
                                df_realty['locality_name']
                                .str.replace('садоводческое некоммерческое товарищество','садовое товарищество')
)
```

```python
len(df_realty['locality_name'].unique())
```

Количество уникальных значений изменилось с <b>364</b> до <b>318</b>. Дальнейшие замены (например, убирать "тип" населенного пункта) могут привести к ошибкам, т.к. могут существовать город и поселок с одинаковыми названиями.

Нет возможности определить, где была допущена ошибка. Объем данных незначителен (**64** объекта), удалим эти данные.

```python
df_realty = df_realty[~((df_realty['living_area'] + df_realty['kitchen_area']) > df_realty['total_area'])]
```

Попробуем заменить пустые значения:

Для объектов, у которых отсутствует и жилая площадь, и площадь кухни: медианный процент каждой из площадей от общей

```python
kitchen_area_part =  (df_realty['kitchen_area'] / df_realty['total_area']).median()
living_area_part =  (df_realty['living_area'] / df_realty['total_area']).median()
print(kitchen_area_part, living_area_part, kitchen_area_part + living_area_part)
```

Сумма долей меньше единицы, это значит, сумма площадей будет точно меньше общей площади помещения.


Произведем замены.

```python
bool_array = (df_realty['kitchen_area'].isna()) & (df_realty['living_area'].isna())
df_realty.loc[bool_array,'kitchen_area'] = round(kitchen_area_part * (df_realty['total_area']),2)
df_realty.loc[bool_array,'living_area'] = round(living_area_part  * (df_realty['total_area']),2)
```

Для объектов, у которых отсутствует информацию про жилую площадь, но есть площадь кухни - возьмем медианный процент из площади помещения за вычетом площади кухни

```python
area_part = (df_realty['living_area'] / (df_realty['total_area'] - df_realty['kitchen_area'])).median()
bool_array = (df_realty['kitchen_area'].notna()) & (df_realty['living_area'].isna())
df_realty.loc[bool_array,'living_area'] = round(area_part * (df_realty['total_area'] - df_realty['kitchen_area']),2)
```

Аналогичную замену сделаем для объектов, где есть информация про жилую площадь и нет про площадь кухни.

```python
area_part = (df_realty['kitchen_area'] / (df_realty['total_area'] - df_realty['living_area'])).median()
bool_array = (df_realty['kitchen_area'].isna()) & (df_realty['living_area'].notna())
df_realty.loc[bool_array,'kitchen_area'] = round(area_part * (df_realty['total_area'] - df_realty['living_area']),2)
```

**Используемые функции**

```python
def base_analyse(name_column: str,
                 data: pd.DataFrame = df_realty,
                 show_missing: bool = True,
                 show_correlation: bool = True,
                 show_boxplot: bool = True,
                 is_bar: bool = False, 
                 plt_rot: int = 0,
                 hist_bins: int = 0) -> None: 
    if show_missing:
        if any(data[name_column].isna()):
            print(f"Количество пропущенных значений {data[name_column].isna().sum()}")
            print(f"Процент пропущенных значений {data[name_column].isna().mean()}")
    describ = data[name_column].describe()
    describ['10%'] = data[name_column].quantile(0.1)
    describ['90%'] = data[name_column].quantile(0.9)
    if hist_bins == 0:
        hist_bins = int(describ['max'] - describ['min'] + 1)
    if is_bar:
        data[name_column].value_counts().sort_index().plot.bar(rot=plt_rot)
    else:
        data[name_column].hist(bins=hist_bins, xrot=plt_rot)
    plt.title(title_dict.get(name_column, name_column))
    plt.show()
    print(describ)
    if show_boxplot:
        sns.boxplot(data=data[name_column], orient='h', color='#D0F0C0', showmeans=True)
        plt.title(title_dict.get(name_column, name_column))
        plt.show()
    if show_correlation:
        print(f"Корреляция между {name_column} и общей стоимостью объекта: {data['last_price'].corr(data[name_column]):.2f}")
        print(f"Корреляция между {name_column} и ценой квадратного метра: {data['cost_per_meter'].corr(data[name_column]):.2f}")
        
```

```python
def correlation_analyse(name_column: str,
                        data: pd.DataFrame = df_realty,
                        correlation_with: str = 'last_price',
                        is_hexbin: bool = False,
                        is_categorial: bool = False,
                        show_boxplot: bool = True,
                        show_corr: bool = True,
                        scatter_aplha: int = 1,
                        hexbin_gridsize: int = 20) -> None: 
    
    if  not is_categorial:
        if is_hexbin:
            data.plot(x=name_column, 
                      y=correlation_with, 
                      kind='hexbin', 
                      gridsize=hexbin_gridsize,
                      sharex=False, 
                      grid=True)
        else:
            data.plot(x=name_column, 
                      y=correlation_with,           
                      kind='scatter',
                      alpha=scatter_aplha)
        plt.title(title_dict.get(name_column, name_column))        
        plt.show()
        if show_corr:
            print(f"Коэффициент корреляции {data[correlation_with].corr(data[name_column]):.1f}")
    else:
        data.pivot_table(index=name_column, 
                         values=correlation_with, 
                         aggfunc=['median', 
                                  lambda x: x.quantile(0.75) - x.median()]).plot.bar(stacked=True)
        plt.legend(['median','75%'])
        plt.title(title_dict.get(name_column, name_column))
        plt.show()
        if show_boxplot:
            sns.boxplot(data=data, y=name_column, x=correlation_with, orient='h')
            plt.title(title_dict.get(name_column, name_column))
            plt.show()
        
```
